```python
import splines as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_weights = 10
w = np.random.randn(n_weights)
print(w.shape)
x = np.arange(0, 15, 0.05)
print(x.shape, x.min(), x.max())
s = sp.spline(w, x.min(), x.max())
y = s(x)
print(y.shape)
```

```python
plt.plot(x, y)
plt.show()
```

```python
# o importante aqui é o scale (desvio padrão do ruído)
ruido = np.random.normal(loc=0.0, scale=.5, size=x.shape[0])
# vamos tentar recuperar um cosseno!
y = np.cos(x) + ruido

plt.scatter(x,y)
plt.show()
```

```python
M = np.zeros((x.shape[0], n_weights))
for i in range(x.shape[0]):
    for j in range(n_weights):
        M[i,j] = s.beta_j(j, x[i])
M
```

```python
M2 = sp.matrix_m2(n_weights)
M2
```

```python
b = M.T.dot(y)
b
```

```python
M1 = M.T.dot(M) - M2
M1
```

```python
lamb = 1
w = np.linalg.solve(M1 + lamb*M2, b)
w
```

```python
s = sp.spline(w, x.min(), x.max())
plt.scatter(x,y)
plt.plot(x, s(x), c='r', linewidth=2)
plt.show()
```
